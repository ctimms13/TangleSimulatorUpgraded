# Different rates

```python
### Chain 20, different rates
r1 = pd.read_csv("pc-r1-c20 100")
r3 = pd.read_csv("pc-r3-c20 100")
r5 = pd.read_csv("pc-r5-c20 100")
times = []

Nr1 = pd.read_csv("Non-PC-r1 100")
Nr3 = pd.read_csv("Non-PC 100")
Nr5 = pd.read_csv("Non-PC-r5 100")

for n in range(1,61):
    times.append(n)

PCR1 = r1['1'].tolist()[:60]
PCR3 = r3['1'].tolist()[:60]
PCR5 = r5['1'].tolist()[:60]

NPCR1 = Nr1['1'].tolist()[:60]
NPCR3 = Nr3['1'].tolist()[:60]
NPCR5 = Nr5['1'].tolist()[:60]

plt.plot(times, PCR1, label='Chain: Rate = 1', color='b')
plt.plot(times, NPCR1, label='No Chain: Rate = 1', color='b', linestyle='dashed')
plt.plot(times, PCR3, label='Rate = 3', color='g')
plt.plot(times, NPCR3, label='No Chain: Rate = 3', color='g', linestyle='dashed')
plt.plot(times, PCR5, label='Rate = 5', color='r')
plt.plot(times, NPCR5, label='No Chain: Rate = 5', color='r', linestyle='dashed')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Number of confirmations')
```
